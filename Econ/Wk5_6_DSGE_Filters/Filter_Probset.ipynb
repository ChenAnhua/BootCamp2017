```python
import pandas_datareader.data as web
import datetime
import scipy.signal as sig
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm
```

```python
start = datetime.datetime(1960, 1, 1)
end = datetime.datetime(2015, 12, 1)
GDP = web.DataReader("GDPC1", "fred", start, end) 
CONSUMP = web.DataReader("PCECC96", "fred", start, end)
INV = web.DataReader("GPDIC1", "fred", start, end)
GDPDEF = web.DataReader("GDPDEF", "fred", start, end)
#taking log
logGDP = np.log(GDP.values)
logCONSUMP = np.log(CONSUMP.values)
logINV = np.log(INV.values)
logGDPDEF = np.log(GDPDEF.values)
GDPfreq, GDPper = sig.periodogram(logGDP, axis=0)
CONSUMPfreq, CONSUMPper = sig.periodogram(logCONSUMP, axis=0)
INVfreq, INVper = sig.periodogram(logINV, axis=0)
GDPDEFfreq, GDPDEFper = sig.periodogram(logGDPDEF, axis=0)
#plotting
plt.plot(GDPfreq[1:-1], np.log(GDPper[1:-1]))
plt.title('GDP')
plt.xlabel('frequency')
plt.show()
plt.plot(CONSUMPfreq[1:-1], np.log(CONSUMPper[1:-1]))
plt.title('Consumption')
plt.xlabel('frequency')
plt.show()
plt.plot(INVfreq[1:-1], np.log(INVper[1:-1]))
plt.title('Investment')
plt.xlabel('frequency')
plt.show()
plt.plot(GDPDEFfreq[1:-1], np.log(GDPDEFper[1:-1]))
plt.title('GDP deflator')
plt.xlabel('frequency')
plt.show()
```
